```python
def grader(score):
    if score >= 87:
        print(f'You got {score} points! That is an A!')
    elif score >= 82:
        print(f'You got {score} points! That is a B!')
    elif score >= 72:
        print(f'You got {score} points! That is a C!')
    elif score >= 64:
        print(f'You got {score} points! That is a D!')  
    else: 
        print(f'You got {score} points! That is an F!')
```
