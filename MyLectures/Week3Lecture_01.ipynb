```

```

```

```

So far we've only been using `where` with the predicate that requires finding the values in a column to be *exactly* equal to a certain value. However, there are many other predicates. Here are a few:

|Predicate|Example|Result|
|-|-|-|
|`are.equal_to`|`are.equal_to(50)`|Find rows with values equal to 50|
|`are.not_equal_to`|`are.not_equal_to(50)`|Find rows with values not equal to 50|
|`are.above`|`are.above(50)`|Find rows with values above (and not equal to) 50|
|`are.above_or_equal_to`|`are.above_or_equal_to(50)`|Find rows with values above 50 or equal to 50|
|`are.below`|`are.below(50)`|Find rows with values below 50|
|`are.between`|`are.between(2, 10)`|Find rows with values above or equal to 2 and below 10|
|`are.between_or_equal_to`|`are.between_or_equal_to(2, 10)`|Find rows with values above or equal to 2 and below or equal to 10|

````python
# records with SALARY above 10 million
nbax = nba2015.where('SALARY', are.above(10)).show()
nbax
````

````python
# salary above 10 million then sorted in ascending order
nba2015.where('SALARY', are.above(10)).sort('SALARY').show()
````

````python
# records with PLAYER 'Stephen Curry'
nba2015.where('PLAYER', are.equal_to('Stephen Curry'))
````

````python
# records with Golden State Warriors
nba2015.where('TEAM', are.equal_to('Golden State Warriors')).show()
````

````python
# records with Team Denver Nuggets
nba2015.where('TEAM', are.equal_to('Denver Nuggets')).show()
````

````python
# Point Guard with salary above 15 million
nba2015.where('POSITION', are.equal_to('PG')).show()
````

````python
# salary between 10 and less than 10.3
nba2015.where('SALARY', are.between(10, 10.3))
````

````python
# Player Barack Obama
nba2015.where('PLAYER', are.equal_to('Barack Obama'))
````

````python
# records containing Warriors
nba2015.where('TEAM', are.containing('Warriors'))     
````
