# Changen_toint function to change string to integer

```python
# Make sure you run this function only once otherwise; an error will occur
def changen_toint(array_data, df):
    for x in array_data:
        y = df[x]    
        u = []        
        for i in range(0, len(y)):
            z = y[i] 
            
            if z != '?':
                if z.find('.') == -1:
                    w = int(z)              
                    u.append(w)
                else:
                    u.append(pd.NA)
            else:
                    u.append(pd.NA)
                
           
        df[x] = u
```
